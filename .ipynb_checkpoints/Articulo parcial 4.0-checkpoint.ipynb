### creamos metodos y variables necesarios mas adelante

```python
imagenes = []
ojo_derecho_escalado = []
ojo_izquierdo_escalado =[]
boca_escalada =[]
imagenes_escaladas = []
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
mouth_cascade = cv2.CascadeClassifier('haarcascade_mcs_mouth.xml')
```

```python


def cargar_imagenes():
    imagenes_cargadas = []
    for i in range (1,600,1):
        imagen = cv2.imread(f'ImagenesMelas/file{i}.jpg')
        if imagen is not None:
            imagenes_cargadas.append(imagen)
    for i in range (3400,4000,1):
        imagen = cv2.imread(f'ImagenesMelas/file{i}.jpg')
        if imagen is not None:
            imagenes_cargadas.append(imagen)
    return imagenes_cargadas

def aplicar_escala_de_grises_a_imagen(foto):
        return cv2.cvtColor(foto, cv2.COLOR_BGR2GRAY)
    
def aplicar_filtro_gaussiano(imagen):
    return cv2.GaussianBlur(imagen,(5,5),0)

def aplicar_umbral(imagen):
    return cv2.adaptiveThreshold(imagen, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 199, 5)  

def preparar_imagenes_para_vectorizar(imagenes):
    imagenes_preparadas = []
    for i in imagenes:
        i = aplicar_filtro_gaussiano(i)
        i = aplicar_umbral(i)
        imagenes_preparadas.append(i)
    return imagenes_preparadas
        

def vectorizar_imagenes(imagenes):
    imagenes_vectorizadas =[]
    for i in imagenes:
        imagenes_vectorizadas.append(i.reshape(1,15000))
    return imagenes_vectorizadas

def distancia(x1,y1,x2,y2):
    x=(x2-x1)**2
    y=(y2-y1)**2
    dist=(x+y)**0.5
    return dist

def detectar_caracteristicas_prueba(imagenes):
    numero =1
    contador = 0
    for img in imagenes:
        gray =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(img, 1.35, 1)
        for (x,y,w,h) in faces:
            #Parametros aproximados basados en las proporciones del rostro
            Eyaprox=int(3*h/8)
            Ex1aprox=int(3*w/10)
            Ex2aprox=int(7*w/10)
            mxaprox=int(w/2)
            myaprox=int(4*h/5)

            #Puntos de referencia para ojos y boca
            Ey1=Eyaprox
            Ey2=Eyaprox
            Ex1=Ex1aprox
            Ex2=Ex2aprox
            Ew1=5
            Ew2=5
            Eh1=5
            Eh2=5
            E1Dist=9999
            E2Dist=9999
            Mx=mxaprox
            My=myaprox
            Mw=5
            Mh=5   
            Mdist=9999

            img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,100,100),2)
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = img[y:y+h, x:x+w]
            roi_oe = gray[y:y+h, x:x+w]
            eyes = eye_cascade.detectMultiScale(roi_gray)
            mouth = mouth_cascade.detectMultiScale(roi_oe)
            for(ex,ey,ew,eh) in eyes:
                exactual=ex+int(ew/2)
                eyactual=ey+int(eh/2)
                e1dist=distancia(exactual,eyactual,Ex1aprox,Eyaprox)
                e2dist=distancia(exactual,eyactual,Ex2aprox,Eyaprox)
                if e1dist<E1Dist or e2dist<E2Dist:
                    if e1dist<e2dist:
                        Ex1=ex
                        Ey1=ey
                        Ew1=ew
                        Eh1=eh
                        E1Dist=e1dist
                    else:
                        Ex2=ex
                        Ey2=ey
                        Ew2=ew
                        Eh2=eh
                        E2Dist=e2dist
    
            for(mx,my,mw,mh) in mouth:
                myactual=my+mh/2
                mxactual=mx+mw/2
                mdist=distancia(mxactual,myactual,mxaprox,myaprox)
                if mdist<Mdist:
                    Mx=mx
                    My=my
                    Mw=mw
                    Mh=mh
                    Mdist=mdist
            ojo_derecho = roi_gray[Ey1:Ey1+Eh1, Ex1:Ex1+Ew1]
            ojo_derecho_escalado.append(cv2.resize(ojo_derecho, (150, 100)))
            ojo_izquierdo = roi_gray[Ey2:Ey2+Eh2, Ex2:Ex2+Ew2]
            ojo_izquierdo_escalado.append(cv2.resize(ojo_izquierdo, (150, 100)))
            boca = roi_gray[My:My+Mh, Mx:Mx+Mw]
            boca_escalada.append(cv2.resize(boca, (150, 100)))

        cv2.rectangle(roi_color,(Ex1,Ey1),(Ex1+Ew1,Ey1+Eh1),(150,150,50),2)
        cv2.rectangle(roi_color,(Ex2,Ey2),(Ex2+Ew2,Ey2+Eh2),(150,50,150),2)
        cv2.rectangle(roi_color,(Mx,My),(Mx+Mw,My+Mh),(0,0,255),2)

    
```

```python

ojo_derecho_preparado = np.array(vectorizar_imagenes(preparar_imagenes_para_vectorizar(ojo_derecho_escalado)))
ojo_izquierdo_preparado = np.array(vectorizar_imagenes(preparar_imagenes_para_vectorizar(ojo_izquierdo_escalado)))
boca_preparada = np.array(vectorizar_imagenes(preparar_imagenes_para_vectorizar(boca_escalada)))
```

```python
imagenes_concatenadas = []
for i in range(0,len(ojo_derecho_preparado),1):
    imagenes_concatenadas.append(np.hstack((ojo_derecho_preparado[i],ojo_izquierdo_preparado[i],boca_preparada[i])))
imagenes_concatenadas = np.array(imagenes_concatenadas).reshape(len(ojo_derecho_preparado),45000) 
```

### Creamos un data frame en el cual cada registro representa una imagen vectorizada

```python
columns = [f'c{i}' for i in range(45000)]
data = imagenes_concatenadas 
data_frame = pd.DataFrame(data = data, columns = columns)
```

```python
sonrie = []
for i in range(0,600,1):
    if i<=300:
        sonrie.append(1)
    else:
         sonrie.append(0)         
data_frame['Sonrie'] =sonrie
```

### Separamos datos de test de datos de train

```python
train_set, test_set = train_test_split(data_frame,test_size=0.2,random_state=42)
```

```python
train_set.head()
```

### Creamos variables con x ^ y diferensiando train de test

```python
Xknn_train = np.c_[train_set.iloc[:,0:44999]]
yknn_train = np.r_[train_set['Sonrie']]
Xknn_test = np.c_[test_set.iloc[:,0:44999]]
yknn_test = np.r_[test_set['Sonrie']]
```

### Aplicamos cross validation con una grilla con k vecinos impares para garantizar que no haya empate, y hallar asi un k optimo

```python
grid_params ={
    'n_neighbors':[1,3,5,7]
}

gs = GridSearchCV(
    KNeighborsClassifier(),
    grid_params,
    verbose =1,
    cv=3,
    n_jobs=-1

)

gs_results = gs.fit(Xknn_train,yknn_train)
```

###  Obtenemos el k optimo y entrenamos el modelo


```python
gs_results.best_params_
```

```python
neigh = KNeighborsClassifier(n_neighbors=5)
neigh.fit(Xknn_train,yknn_train);
```

```python
y_predict_train=neigh.predict(Xknn_train)
y_proba_train = neigh.predict_proba(Xknn_train)
```

### Presentacin de resultados 

```python
y_predict_test=neigh.predict(Xknn_test)
y_proba_test = neigh.predict_proba(Xknn_test)
```

```python
acc_train = accuracy_score(yknn_train, y_predict_train)
acc_test = accuracy_score(yknn_test, y_predict_test)
precision_train = precision_score(yknn_train,y_predict_train)
precision_test = precision_score(yknn_test,y_predict_test)
recall_train = recall_score(yknn_train,y_predict_train)
recall_test=recall_score(yknn_test,y_predict_test)
```

```python
print("Test accuracy: %.2f" % acc_test)
print("Traim accuracy: %.2f" % acc_train)
print("Test precision: %.2f" % precision_test)
print("Train precision: %.2f" % precision_train)
print("Test recall: %.2f" % recall_test)
print("Train recall: %.2f" % recall_train)
```

### Representacion de curva ROC

```python
import sklearn.metrics as metrics
probs = neigh.predict_proba(Xknn_test)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(yknn_test, preds)
roc_auc = metrics.auc(fpr, tpr)

import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('Tasa verdaderos positivos')
plt.xlabel('Tasa falsos positivos')
```
